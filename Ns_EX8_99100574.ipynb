# Importing Libraries

```python
import numpy as np 
import matplotlib.pyplot as plt 
from numba import njit 
from copy import deepcopy
import random
```

# Creating 6 Orthogonal Patterns 

```python
c = 1 
@njit
def overlap(a1,a2):
    return np.sum(a1 * a2)

m1 = np.ones((8,8),dtype = np.int64)
m2 = np.ones((8,8),dtype = np.int64)
m3 = np.ones((8,8),dtype = np.int64)
m4 = np.ones((8,8),dtype = np.int64)
m5 = np.ones((8,8),dtype = np.int64)
m6 = np.ones((8,8),dtype = np.int64)
m7 = np.ones((8,8),dtype = np.int64)
m8 = np.ones((8,8),dtype = np.int64)


m1 = 2*(np.random.random((8,8)) >= 0.5) -1 
while overlap(m1,m2) !=0 :
    m2 = 2*(np.random.random((8,8)) >= 0.5) -1
while  overlap(m1,m3) !=0 or  overlap(m2,m3) !=0 :
    m3 = 2*(np.random.random((8,8)) >= 0.5) -1
while  overlap(m1,m4) !=0 or  overlap(m2,m4) !=0 or  overlap(m3,m4) !=0  :
    m4 = 2*(np.random.random((8,8)) >= 0.5) -1
while  overlap(m1,m5) !=0 or  overlap(m2,m5) !=0 or  overlap(m3,m5) !=0 or  overlap(m4,m5) !=0  :
    m5 = 2*(np.random.random((8,8)) >= 0.5) -1
while  overlap(m1,m6) !=0 or  overlap(m2,m6) !=0 or  overlap(m3,m6) !=0 or  overlap(m4,m6) !=0 or  overlap(m5,m6) !=0   :
    m6 = 2*(np.random.random((8,8)) >= 0.5) -1
while  overlap(m1,m7) !=0 or  overlap(m2,m7) !=0 or  overlap(m3,m7) !=0 or  overlap(m4,m7) !=0 or  overlap(m5,m7) !=0 or  overlap(m6,m7) !=0  :
    m7 = 2*(np.random.random((8,8)) >= 0.5) -1
while  overlap(m1,m8) !=0 or  overlap(m2,m8) !=0 or  overlap(m3,m8) !=0 or  overlap(m4,m8) !=0 or  overlap(m5,m8) !=0 or  overlap(m6,m8) !=0  or  overlap(m7,m8) !=0  :
    m8 = 2*(np.random.random((8,8)) >= 0.5) -1

print(overlap(m1,m2))
print(overlap(m1,m3))
print(overlap(m1,m4))
print(overlap(m1,m5))
print(overlap(m1,m6))
print(overlap(m1,m7))
print(overlap(m1,m8))
print(overlap(m2,m3))
print(overlap(m2,m4))
print(overlap(m2,m5))
print(overlap(m2,m6))
print(overlap(m2,m7))
print(overlap(m2,m8))
print(overlap(m3,m4))
print(overlap(m3,m5))
print(overlap(m3,m6))
print(overlap(m3,m7))
print(overlap(m3,m8))
print(overlap(m4,m5))
print(overlap(m4,m6))
print(overlap(m4,m7))
print(overlap(m4,m8))
print(overlap(m5,m6))
print(overlap(m5,m7))
print(overlap(m5,m8))
print(overlap(m6,m7))
print(overlap(m6,m8))
print(overlap(m7,m8))

```

```python
eta1 = m1.reshape((-1,1))
eta2 = m2.reshape((-1,1))
eta3 = m3.reshape((-1,1))
eta4 = m4.reshape((-1,1))
eta5 = m5.reshape((-1,1))
eta6 = m6.reshape((-1,1))
eta7 = m7.reshape((-1,1))
eta8 = m8.reshape((-1,1))

w1 = (eta1 @ (eta1.T))
w2 = (eta2 @ (eta2.T))
w3 = (eta3 @ (eta3.T))
w4 = (eta4 @ (eta4.T))
w5 = (eta5 @ (eta5.T))
w6 = (eta6 @ (eta6.T))
w7 = (eta7 @ (eta7.T))
w8 = (eta8 @ (eta8.T))

w_net = c * (w1 + w2 + w3 + w4 + w5 + w6 + w7 + w8)

plt.imshow(w_net)
```

# Determenistic Dynamic

```python
def step(state, w):
    v_state = state.reshape((-1,1))
    v_new_state = 2* ((w @ v_state) >= 0 ) - 1 
    new_state = v_new_state.reshape((8,8))
    return new_state

    
```

## the stochastic dynamic

```python
def g(h,beta):
    return 0.5 * (1 + np.tanh(beta*h))
```

```python
def stochastic_step(state, w,beta):
    v_state = state.reshape((-1,1))
    v_new_state = 2*(np.random.random((64,1)) <= g(w @ v_state , beta)) -1 
    new_state = v_new_state.reshape((8,8))
    return new_state

    
```

## the stichastic effect with beta = 0.02

```python
state = deepcopy(m4)
plt.imshow(state)
plt.show()
for i in range(1000):
    state = stochastic_step(state , w_net,0.02)

print(overlap(state , m1))
print(overlap(state , m2))
print(overlap(state , m3))
print(overlap(state , m4))
print(overlap(state , m5))
print(overlap(state , m6))
print(overlap(state , m7))
print(overlap(state , m8))

plt.imshow(state)
plt.show()
```
