```python
import numpy as np 
import matplotlib.pyplot as plt 
```

# defining box muler algorithem 

```python
# box muler algorithem: 
# y = sqrt(-2log(r1)).cos(2pi.r2)
# r1 and r2 from uniform dist. between 0 and 1 
def getGaussian(rs):
    r1 = rs[0,:]
    r2 = rs[1,:]
    return np.sqrt(-2*np.log(r1))*np.cos(2*np.pi*r2)
```

# testing box muler implementation 

```python
steps = 10000
rs = np.random.uniform(0,1,(2,steps))
y = getGaussian(rs)

plt.title("uniform distrobution")
plt.hist(rs.reshape((-1,1)))
plt.show()

plt.title("gaussian distrobution")
plt.hist(y)
plt.show()

# check mean and std of generated data 
print("mean: ",np.mean(y),"std: ", np.std(y))
```

# I0 = 0.8, thersolrd = 1 , u0 = 0 , dt = 0.001 , sigma = 1 , steps = 1e6 

```python
steps = 1_000_000
rs = np.random.uniform(0,1,(2,steps))
y = getGaussian(rs)

u = 0 
I0 = 0.8
dt = 0.001
sigma = 1
thershold = 1 
data = np.zeros(steps)
for i in range(steps):
    data[i] = u
    du = (-u + I0)*dt + sigma * np.sqrt(dt) * y[i]
    u += du 
    if u > thershold :
        u = 0 

plt.title("overal potential behaviar plot")
plt.plot(np.arange(steps)*dt, data)
plt.show()

plt.title("zoomed potential behaviar plot for first 300 steps")
plt.plot(dt*np.arange(300), data[:300])
plt.show()

plt.title("IRI distribution")
plt.hist(dt*np.diff(np.argwhere(data == 0)[:,0]),bins = 20,density = True)
plt.show()
```
