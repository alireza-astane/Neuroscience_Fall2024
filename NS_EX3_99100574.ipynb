```python
xs = np.random.uniform(0,1,10000)
```

```python
plt.hist(xs)
```

```python
def P(x,q0):
    return -q0*np.log(1-x)
```

```python
fs = P(xs,1)
```

```python
plt.hist(fs)
plt.savefig("f1.png")
```

```python
steps = 10000
dt = 0.001


def getU(u,q):
    newU =  u*(1 - dt) + q
    return np.where(newU > 1, 0 , newU)

def plot(u0,q0,powerSteps):
    steps = 10**powerSteps
    uData= np.zeros(steps)
    xs = np.random.uniform(0,1,steps)
    fs = P(xs,q0)
    u = u0
    
    for i in range(steps):
        uData[i] = u 
        u = getU(u,fs[i])
    plt.plot(np.arange(steps), uData)
    return()
```

```python
plot(0,0.00091,5)
plt.savefig("f2.png")
```
